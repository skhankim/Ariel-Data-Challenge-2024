## Markov Chain

현재 상태가 미래에만 영향을 끼치고 과거의 상태와는 독립적이라는 **Markov 성질**을 가지고 있어 현재 상태만으로 미래 상태를 예측할 수 있는 확률적 과정
    
    
Markov  체인은 **상태 공간** 위에서만 정의된다.
    

현재 상태에서 다음 상태로 이동할 확률을 전이 확률이라 하며 이 확률이 시간에 따라 일정한 경우 정상 Markov  체인이라 한다.

강한 Markov  체인의 경우 일반적인 Markov  체인을 유연하게 확장한 개념으로 어떤 조건이 만족되는 시점을 의미하는 정지 시간 $\tau$에 도달했을 때 그 이후 시점 t에서의 상태 $X_{\tau+t}$는 오직 정지 시간에서의 상태 $X_\tau$에만 조건화되는 것을 의미한다.

엄밀한 수학적 증명을 통해 특정 확률 과정들이 Markov  성질을 갖는다는 경우를 확인하는 경우도 있지만 대개 우선적으로 Markov  성질을 가지고 있다는 가정 하에 Markov  체인을 적용하여 결과 분석을 통해 Markov 성질을 가지고 있다고 판단하는 경우가 많다.

이런 Markov 체인 성질을 이용하여 Baysian Inference의 가장 계산이 복잡한 부분인 사후 확률을 계산하는 공식의 분모에 해당하는 Evidence를 구하는 난점을 쉽게 해결할 수 있기 때문에 Baysian Infrence와 관련된다.

따라서 Markov 체인 성질을 갖는 확률 분포의 경우 시간이 충분히 흐르고 나면 이론적으로 무조건 stationary distribution에 수렴하는 성질을 가지고 있으므로 이런 성질을 이용한 Sampling 방법으로 Monte Carlo Markov Chain이 있다.

확률 변수가 한 두 개가 아니고 여러 개로 구성되어 있고 심지어 각 확률 변수가 서로 독립이 아니라 상호 의존성을 가지고 복잡하게 얽혀 있는 고차원 확률 분포의 경우 어떤 변수의 값이 다른 변수들의 값에 영향을 미치므로 특정 변수의 값이 결정되면 다른 변수들의 값이 이에 따라 좁혀져 단순히 독립적인 난수를 생성하여 샘플링 할 수도 없고 전체 결합 분포를 수식으로 명시적으로 표현하거나 직접 계산하기 어렵다. 따라서 Markov Chain을 통해 점진적으로 Target 분포에 수렴하도록 설계하여 최종 Target 분포를 근사하여 Sampling할 수 있고 이를 통해 기대 값, 분산, 확률 밀도 등 다양한 특성을 추정할 수 있다. 

이에 대한 대표적인 알고리즘으로 Metropolis-Hastings Algorithm과 Gibbs Sampling 방법이 있다.

- **Metropolis-Hastings Algorithm**: 제안 분포에서 새로운 상태를 생성하고, 그 상태를 채택할 확률을 조건부 확률에 따라 결정하는 방식이다.
- **Gibbs Sampling**: 다차원 분포에서 각 변수의 조건부 분포를 이용해 순차적으로 Sampling하는 방법이다.

두 방법 모두 Markov  체인을 통해 샘플을 생성하며, 충분히 많은 샘플을 생성하면 목표 분포에 가까운 샘플을 얻을 수 있다.

## 용어 설명

### Markov 성질
> 확률 공간 (Ω,F,P)가 여과 (Fs,s∈I) (일정하게 순서가 있는 지표 집합 I)를 가지고 있고, (S,$S$)가 측도 공간이라고 하자. (S,$S$)-값을 가지는 확률 과정 X=${X_t:Ω→S}_{t∈I}$가 여과에 적응되어 있을 때, 각 A∈$S$와 각 s,t∈I에 대해 s<t이면,
> 
- **$P(X_t∈A∣F_s) = P(X_t∈A∣X_s)$**
가 성립할 때, X가 Markov 성질을 갖는다고 한다. 이는 이전의 과거들에 대한 누적된 정보에서 확률 변수 $X_t$가 사건 $A$에 속할 확률과 시간 s에서의 상태 $X_s$만 알고 있을 때 미래 상태가 사건 A에 속할 확률이 같다는 것을 의미한다. 
    
    1. F는 사건 집합으로 $\Omega$의 부분 집합들로 구성된 sigma algebra로 측정 가능한 모든 사건을 포함하여, 확률을 할당할 수 있도록 하는데, 여기서 측정 가능한 모든 사건이란 
    
    1. 공집함 포함: Ω의 공집합(∅)은 F에 포함된다.→아무 일도 일어나지 않는 사건, 기초가 된다.
    2. 여집합에 대해 닫혀 있음: $A ∈ F$이면, 그 여집합 $A^c$도 F에 속합니다.→사건이 일어날 확률 뿐 아니라, 일어나지 않을 확률도 정의할 수 있어야 함.
    3. 가산 합집합에 대해 닫혀 있음: A₁, A₂, A₃, ⋯ ∈ F라면, 이들의 합집합 ⋃ᵢ₌₁^∞ Aᵢ도 F에 속한다.→무한 개의 사건들이 동시에 일어날 수 있는 가능성에 대한 확률을 정의할 수 있어야 함
    
    위 세가지 닫힘 조건을 만족하는 사건들로 계산의 복잡성을 최소화하는 측정 가능한 사건들의 최소 집합으로써 sigma algebra를 구성할 수 있다.
    

가 성립할 때, X가 Markov 성질을 갖는다고 한다. 이는 이전의 과거들에 대한 누적된 정보에서 확률 변수 $X_t$가 사건 $A$에 속할 확률과 시간 s에서의 상태 $X_s$만 알고 있을 때 미래 상태가 사건 A에 속할 확률이 같다는 것을 의미한다.

### $P(X_t∈A∣F_s) = P(X_t∈A∣X_s)$
- F는 사건 집합으로 $\Omega$의 부분 집합들로 구성된 sigma algebra로 측정 가능한 모든 사건을 포함하여, 확률을 할당할 수 있도록 하는데, 여기서 측정 가능한 모든 사건이란 
    
    1. 공집함 포함: Ω의 공집합(∅)은 F에 포함된다.→아무 일도 일어나지 않는 사건, 기초가 된다.
    2. 여집합에 대해 닫혀 있음: $A ∈ F$이면, 그 여집합 $A^c$도 F에 속합니다.→사건이 일어날 확률 뿐 아니라, 일어나지 않을 확률도 정의할 수 있어야 함.
    3. 가산 합집합에 대해 닫혀 있음: A₁, A₂, A₃, ⋯ ∈ F라면, 이들의 합집합 ⋃ᵢ₌₁^∞ Aᵢ도 F에 속한다.→무한 개의 사건들이 동시에 일어날 수 있는 가능성에 대한 확률을 정의할 수 있어야 함
    
    위 세가지 닫힘 조건을 만족하는 사건들로 계산의 복잡성을 최소화하는 측정 가능한 사건들의 최소 집합으로써 sigma algebra를 구성할 수 있다.
- P는 확률 측도로 F에 있는 각 사건에 확률을 할당하는 함수. 즉, 특정 사건이 일어날 확률을 나타낸다.
- 여과($F_s$)는 시간 s까지 알고 있는 정보의 누적된 집합을 나타내는 sigma algebra의 모음으로 시간이 흐르면서 누적된 정보를 의미한다.
- 지표 집합(I)는 시간 지표를 나타내는 집합으로 시간의 흐름에 따라 확률 과정이 어떻게 변하는 지를 나타내는 기준이다.
- $S$는 상태 공간 S에 정의된 sigma algebra로 S에 속하는 각 상태에 확률을 할당하기 위한 기준을 제공한다.
- $X_t$는 각 시간 t에 대해 정의된 확률 변수이다.
- $X$는 여러 시간에 걸쳐 정의된 확률 변수들의 집합으로 이를 통해 시간이 지남에 따라 상태가 어떻게 변하는 지를 추적할 수 있다.
- $A$는 상태 공간 S에서 발생할 수 있는 하나의 사건이다.

### 상태 공간
무작위 실험의 모든 가능한 결과들의 집합인 표본 공간 $\Omega$와 달리 시간에 따라 변화할 수 있는 상태들의 집합을 상태 공간이라 한다. 따라서 시스템의 상황에 따라 동적으로 변화할 수 있다.
